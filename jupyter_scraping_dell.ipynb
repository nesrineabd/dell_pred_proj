# My own Dell website scraper with Beautifulsoup

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
from selenium.common.exceptions import WebDriverException
import pandas as pd
```

## 1- Install the right version of the WebDriver 

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

### Function that gets the html of a given html

```python
def get_html(url):
    try:
        driver.get(url)
    except WebDriverException:
        print("page down")
    return driver.page_source
```

### Function that gets the elements needed for data collection

```python
def scrape_data(card):
    _title = card.h3.text.strip()
    _rating = card.find("span", {"class": "ratings-text"}).text
    try:
        price_div = card.find("div", {"class": "ps-dell-price ps-simplified"})
        _price = price_div.select('div > span')[1].get_text(strip=True).strip('€')
    except:
        _price = ''

    first_big_div = card.find("div", {"class": "ps-short-specs ps-iconography" })
    _processeur = first_big_div.select('div > div')[0].get_text(strip=True)
    _windows_version = first_big_div.select('div > div')[1].get_text(strip=True)
    _carte_graphique = first_big_div.select('div > div')[2].get_text(strip=True)
    _RAM = first_big_div.select('div > div')[3].get_text(strip=True)
    try: 
        _disq = first_big_div.select('div > div')[4].get_text(strip=True)
    except:
        _disq = ''
    second_big_div = card.find("div", {"class": "ps-features-icon"})
    _ecran = first_big_div.select('div > div')[0].get_text(strip=True)
    _poids = first_big_div.select('div > div')[1].get_text(strip=True)
    data = {
        'Title': _title,
        'Rate' : _rating,
        'Processeur' : _processeur,
        'Windows Version' : _windows_version,
        'Carte Graphique' : _carte_graphique,
        'RAM' : _RAM,
        'Disque Dur' : _disq,
        'Ecran' : _ecran,
        'Poids' : _poids,
        'Price': _price,
    }
    
    print(data)
    print('\n ')
    return data
```

```python
ads_data = []
num_page = 14
for i in range(1, num_page + 1):
    url = f'https://www.dell.com/fr-fr/search/ordinateur%20portable?p={i}&t=Product&c=laptops&f=true&u=dhs'
    html = get_html(url)
    soup = BeautifulSoup(html, 'lxml')
    cards = soup.find_all('section', {'class': 'ps-top'})
    for card in cards:
        data = scrape_data(card)
        ads_data.append(data)
```

### Put the data in a dataframe and return the csv file

```python
data_frame = pd.DataFrame(ads_data)
data_frame.to_csv('dell_research_results.csv')
data_frame
```
